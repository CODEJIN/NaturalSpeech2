```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '0'
from Inference import Inferencer
import IPython.display as ipd
```

```python
inferencer = Inferencer(
    hp_path= './VCTK_Test/Hyper_Parameters.yaml',
    checkpoint_path= './VCTK_Test/VCTK_3006K.pt',
    batch_size= 8
    )
```

```python
texts = [    
    'No pain, No gain.',
    'Don’t kill the goose that lays the golden eggs.',
    'If you laugh, blessings will come your way.'
    ]
references= [
    'Inference_Wav/p279_003_mic2.flac',
    'Inference_Wav/s5_004_mic2.flac',
    'Inference_Wav/p362_002_mic1.flac',
    ]
ddim_steps = 50
```

```python
audios = inferencer.Inference_Epoch(
    texts= texts,
    references= references,
    ddim_steps= ddim_steps
    )

for text, reference, audio in zip(texts, references, audios):
    print(text)
    print(reference)
    ipd.display(ipd.Audio(audio, rate= inferencer.hp.Sound.Sample_Rate))
    print()
```
