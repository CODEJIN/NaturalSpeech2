```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = '0'
from Inference import Inferencer
import IPython.display as ipd
```

```python
inferencer = Inferencer(
    hp_path= 'Hyper_Parameters.yaml',
    checkpoint_path= './results/LJ_230515/Checkpoint/S_6046.pt',
    batch_size= 8
    )
```

```python
texts = [    
    'No pain, No gain.',
    'Don’t kill the goose that lays the golden eggs.'
    ]
references= ['Inference_Wav/p279_003_mic2.flac'] * len(texts)
ddim_steps = 100
```

```python
audios = inferencer.Inference_Epoch(
    texts= texts,
    references= references,
    ddim_steps= ddim_steps
    )

for text, reference, audio in zip(texts, references, audios):
    print(text)
    print(reference)
    ipd.display(ipd.Audio(audio, rate= inferencer.hp.Sound.Sample_Rate))
    print()
```
